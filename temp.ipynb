```python

path = "temp.xml"

with open(path, "r") as file:
    data = file.read()

print(data)
```

```python
from xml.etree import ElementTree as ET

root = ET.fromstring(data)
namespaces = {
    "tei": "http://www.tei-c.org/ns/1.0"
}  # Add more if there are other namespaces
terms = root.findall(".//tei:term", namespaces=namespaces)

special_tokens = []

# Printing out the text content of each 'term' element
for term in terms:
    special_tokens.append("<" + term.text + ">")

# " ".join(special_tokens[2:])
# special_tokens[2:]

# every other token is a special token
print(special_tokens[2::2])
```
